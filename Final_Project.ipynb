# DATA SET CLEANING
### After the data analysis the data set while need some of the data types to be changed and simplified. Additionally, there were no significant outliers that need to be changed and it is a very large data set. The column names are formatted nicely and a lot of the NaN data types are in important positions. in any of the fuel consumption columns, I plan to replace the NaN with 0 for electric, hybrid, and ethanol fuel type vehicles because the way the data set is designed, they have NaN values. I also plan to remove any duplicates in the data set if there are any because of the large size. Any car that doesn't have a fuel type, mileage, transmission type, or price is going to be removed from the data set. Additionally, since these are listing on the car market, I plan to remove most if not all other NaN values because the likely-hood of the data being incorrect with an NaN on a car listing is very high because there could be an issue with the seller.(if someone left out the registration information or mileage on a car on facebook marketplace or carfax, you would probably be concerned with the validity of the information they give you)

### Remove the unnamed column

### drops first unnamed row of the column

```python
German_Car_Data = German_Car_Data.iloc[:, 1:]

German_Car_Data.head()
```

### Looks at aTotal null value counts in the data set

```python
German_Car_Data.isnull().sum()
```

### turns any electric hybrid or ethanol vehicle's fuel consumption to 0 if the data is not present because it is not present in these rows

```python
for item, row in German_Car_Data.iterrows():
    if pd.isnull(row['fuel_consumption_l_100km']) and row['fuel_type'] in ['Electric', 'Hydrogen', 'Ethanol']:
        German_Car_Data.at[item, 'fuel_consumption_l_100km'] = float(0)

display(German_Car_Data)
```

### Excludes any rows with the fuel type columns as other or unkown because that data is less useful for the analysis

```python
German_Car_Data = German_Car_Data[(German_Car_Data['fuel_type'] != 'Unknown') & (German_Car_Data['fuel_type'] != 'Other')]

display(German_Car_Data)
```

### Goes through each row of fuel_consumption_l_100km and changes it to floats so the data can be used

```python
for index, row in German_Car_Data.iterrows():
    value = row['fuel_consumption_l_100km']
    if isinstance(value, float):
        continue
    elif 'l/100 km' in str(value):
        value = str(value).replace(' l/100 km', '').replace(',', '.')
        try:
            German_Car_Data.at[index, 'fuel_consumption_l_100km'] = float(value)
        except ValueError:
            German_Car_Data.at[index, 'fuel_consumption_l_100km'] = None
    else:
        German_Car_Data.at[index, 'fuel_consumption_l_100km'] = None

display(German_Car_Data)
```

### Goes through each row of fuel_consumption_g_km and changes it to floats so the data can be used

```python
for index, row in German_Car_Data.iterrows():
    value = row['fuel_consumption_g_km']  # Accessing only the 'fuel_consumption_g_km' column
    if isinstance(value, float):
        continue
    elif 'g/km' in str(value):
        value = str(value).replace(' g/km', '').replace(',', '.')
        try:
            German_Car_Data.at[index, 'fuel_consumption_g_km'] = float(value)
        except ValueError:
            German_Car_Data.at[index, 'fuel_consumption_g_km'] = None
    else:
        German_Car_Data.at[index, 'fuel_consumption_g_km'] = None

display(German_Car_Data)
```

### gets rid of any columns in fuel consumption l 100km that are not a float

```python
rows_to_keep = []

for index, row in German_Car_Data.iterrows():
    if isinstance(row['fuel_consumption_l_100km'], float):
        rows_to_keep.append(index)

German_Car_Data = German_Car_Data.loc[rows_to_keep]

display(German_Car_Data)
```

### Because this dataset is so large and is mostly input from users, drop na to make sure that all of the data available is clean data that will help me analyze the dataset. Also changes some row types to the proper types

```python
German_Car_Data.dropna(inplace=True)

German_Car_Data['price_in_euro'] = German_Car_Data['price_in_euro'].astype(int)

German_Car_Data['power_kw'] = German_Car_Data['power_kw'].astype(int)

German_Car_Data['power_ps'] = German_Car_Data['power_kw'].astype(int)

German_Car_Data['fuel_consumption_l_100km'] = German_Car_Data['fuel_consumption_l_100km'].astype(float)

German_Car_Data['fuel_consumption_g_km'] = German_Car_Data['fuel_consumption_g_km'].astype(float)

display(German_Car_Data)
```

# Question #1
### Do manual or automatic vehicles get better mileage? Is there a difference, and if so, which car is more gas-efficient?

```python
# Calculate the average fuel consumption for manual and automatic transmission types
avg_fuel_consumption = German_Car_Data.groupby('transmission_type')['fuel_consumption_l_100km'].mean()
print(avg_fuel_consumption[['Manual', 'Automatic']])
```

```python
# Calculate the average fuel consumption for manual and automatic transmission types
avg_fuel_consumption = German_Car_Data.groupby('transmission_type')['fuel_consumption_l_100km'].mean().loc[['Manual', 'Automatic']]

# Create a bar plot for fuel consumption comparison with narrower bars
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_fuel_consumption.index, y=avg_fuel_consumption.values, palette='pastel', width=0.4)  # Adjusted bar width here
plt.title('Average Fuel Consumption by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Average Fuel Consumption (l/100 km)')
plt.show()
```

### Automatic transmissions on average consume slightly more fuel than cars with manual transmissions. It is not a very noticeable amount but if a consumer wanted to minimize their fuel consumption, they should get a car with a manual transmission.

# Question #2
### Is there a fuel type that is the most efficient for gas mileage, such as a Hybrid, Petrol, Diesel, or Petrol?

```python
# Calculate the average fuel consumption for each fuel type
avg_fuel_by_type = German_Car_Data.groupby('fuel_type')['fuel_consumption_l_100km'].mean().sort_values()
print(avg_fuel_by_type)
```

```python
# Average fuel consumption by fuel type
avg_fuel_by_type = German_Car_Data.groupby('fuel_type')['fuel_consumption_l_100km'].mean().sort_values()
plt.figure(figsize=(12, 8))
plt.barh(avg_fuel_by_type.index, avg_fuel_by_type, color='steelblue')
plt.title('Fuel Efficiency by Type (Lower Consumption is Better)')
plt.xlabel('Average Fuel Consumption (l/100 km)')
plt.ylabel('Fuel Type')
plt.show()
```

### Electric and Hydrogen cars have the best and lowest average fuel consumptions, followed by ethanol then hybrid vehicles. CNG vehicles had the highest fuel consumption on average and would get the worst gas mileage out of all of the fuel types.

# Question #4
### Average Price for cars in different ranges of gas mileage (0-10000, 10000-25000, 25000-50000, 50000-100000, 100000-150000,1500000-200000, 200000+)

```python
# Define mileage and calculate the average price within each
mileage = [0, 10000, 25000, 50000, 100000, 150000, 200000, float('inf')]
labels = ['0-10,000 km', '10,001-25,000 km', '25,001-50,000 km', '50,001-100,000 km', '100,001-150,000 km', '150,001-200,000 km', '200,000+ km']
German_Car_Data['mileage_range'] = pd.cut(German_Car_Data['mileage_in_km'], bins=mileage, labels=labels)
avg_price_by_mileage = German_Car_Data.groupby('mileage_range')['price_in_euro'].mean()
print(avg_price_by_mileage)
```

```python
# Average price within each mileage range
avg_price_by_mileage = German_Car_Data.groupby('mileage_range')['price_in_euro'].mean()
plt.figure(figsize=(12, 6))
plt.plot(avg_price_by_mileage.index, avg_price_by_mileage, marker='o', linestyle='-', color='darkgreen')
plt.title('Average Price by Mileage Range')
plt.xlabel('Mileage Range (km)')
plt.ylabel('Average Price in Euro (€)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Looking at the graph and numerical data, the higher mileage the car has, the less expensive the car is to buy on average. Lower mileage cars are generally more expensive.

# Question #5
### Is there a correlation between the car's horsepower (Ps) and its price?

```python
# Scatter plot for Price vs. Horsepower in PS
plt.figure(figsize=(10, 6))
ax2 = sns.regplot(x='power_ps', y='price_in_euro', data=German_Car_Data, scatter_kws={'alpha':0.5}, line_kws={'color': 'blue'})
ax2.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x} PS'))
ax2.yaxis.set_major_formatter(ticker.StrMethodFormatter('€{x}'))
plt.title('Price vs. Horsepower (PS)')
plt.xlabel('Horsepower (PS)')
plt.ylabel('Price in Euro')
plt.ylim(0, 1500000)  # Set y-axis limit to 1,500,000
plt.grid(True)
plt.show()
```

### The graph shows that generally, the cars with higher horsepower are more expensive than cars with lower horsepower through the regression line. You can also see this with the density of the points go up on the y axis as they progress on the x axis
